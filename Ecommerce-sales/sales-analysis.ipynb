```python
import pandas as pd
import numpy as np
```

```python
sales_df = pd.read_csv('archive/Sale Report.csv')
sales_df
```

# Create pivot tables


```python
inventory_by_item_and_size = pd.pivot_table(sales_df, values='Stock', index='Category', columns='Size', aggfunc='count')
inventory_by_item_and_size
```

```python
dress_design_types = pd.pivot_table(sales_df, values='Stock', index='Design No.', columns='Size', aggfunc='count')
dress_design_types
```

# Create bar graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# remove rows that contain Nan
bargraph_no_na_df = sales_df.dropna()

# generate a frequency count of each category
counts = bargraph_no_na_df['Size'].value_counts() 
print('counts;',counts)
print(counts)
# create a bar plot
plt.bar(counts.index, counts.values)
plt.show()


```
